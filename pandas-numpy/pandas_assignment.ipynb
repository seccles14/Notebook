```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

database_red = 'pandas-intro/intro_pandas/data/winequality-red.csv'
df_red = pd.read_csv(database_red, delimiter=';')


database_white = 'pandas-intro/intro_pandas/data/winequality-white.csv'
df_white = pd.read_csv(database_white, delimiter=';')

```

```python
print(df_red['alcohol'].mean())
print(df_white['alcohol'].mean())
```

```python
red = df_red['alcohol'] > df_red['alcohol'].mean()
white = df_white['alcohol'] > df_white['alcohol'].mean()
print(red)
print(white)
```

```python
redj = df_red.join(red, rsuffix = '_high')
```

```python
whitej = df_white.join(white, rsuffix = '_high')
```

```python
pd.concat([df_red, red], axis = 1)
```

```python
pd.concat([df_white, white], axis = 1)
```

```python
redj[redj['alcohol_high'] == False]
```

```python
whitej[whitej['alcohol_high'] == False]
```

```python
red_bin = pd.cut(df_red['fixed acidity'],5)
red_acid = df_red.join(red_bin, rsuffix = '_bin')
red_acid.head()
```

```python
white_bin = pd.cut(df_white['fixed acidity'],5)
white_acid = df_white.join(white_bin, rsuffix = '_bin')
white_acid.head()
```

```python
red_acid[red_acid['fixed acidity_bin'] != red_acid['fixed acidity_bin'].max()]
```

```python
white_acid[white_acid['fixed acidity_bin'] != white_acid['fixed acidity_bin'].max()]
```

```python
pd.pivot_table(redj, values = 'alcohol', columns = 'alcohol_high', index = 'quality')
```

```python
pd.pivot_table(whitej, values = 'alcohol', columns = 'alcohol_high', index = 'quality')
```

```python
pd.pivot_table(red_acid, values = 'alcohol', columns = 'fixed acidity_bin', index = 'quality')
```

```python
pd.pivot_table(white_acid, values = 'alcohol', columns = 'fixed acidity_bin', index = 'quality')
```
